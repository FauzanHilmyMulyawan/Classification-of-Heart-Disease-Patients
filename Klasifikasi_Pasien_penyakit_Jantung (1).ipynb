**Cek outliers dengan IQR**

```python
Q1 = df_num.quantile(0.25) #Menghitung Q1
Q3 = df_num.quantile(0.75) #Menghitung Q2

IQR = Q3 - Q1 #Menghitung IQR

IQR_outliers = df_num[((df_num < (Q1 - 1.5 * IQR)) |(df_num > (Q3 + 1.5 * IQR))).any(axis=1)] #Menyaring data yang memiliki outliers
IQR_outliers #Menampilkan data yang memiliki outliers
```

**List nama kolom dataset numerik**

```python
num_var = ['age','trtbps','chol','thalachh','oldpeak'] #Membuat list variabel dari dataset numerik
```

**Mencari batas bawah dan batas atas dari variabel dataset numerik**

```python
#Menghitung niai batas atas dan bawah dari tiap variabel di dataset numerik
for i in num_var: #for loop
  IQR = df_num[i].quantile(0.75) - df_num[i].quantile(0.25) #Menghitung nilai IQR
  lower_bound = df_num[i].quantile(0.25) - (1.5*IQR) #Menghitung nilai batas bawah
  upper_bound = df_num[i].quantile(0.75) + (1.5*IQR) #Menghitung batas atas
  
  print(i, ":", lower_bound, ",",  upper_bound) #Menampilkan batas atas dan batas bawah
```

```python
df_num.loc[df_num.trtbps > 170.0 , "trtbps"] = 130 #Mengganti nilai outliers dari variabel trtbps dengan nilai median
df_num.loc[df_num.trtbps < 90.0 , "trtbps"] = 130 #Mengganti nilai outliers dari variabel trtbps dengan nilai median

df_num.loc[df_num.chol > 369.75 , "chol"] = 240 #Mengganti nilai outliers dari variabel chol dengan nilai median
df_num.loc[df_num.chol < 115.75 , "chol"] = 240 #Mengganti nilai outliers dari variabel chol dengan nilai median

df_num.loc[df_num.thalachh > 214.75 , "thalachh"] = 153 #Mengganti nilai outliers dari variabel thalachh dengan nilai median
df_num.loc[df_num.thalachh < 84.75 , "thalachh"] = 153 #Mengganti nilai outliers dari variabel thalachh dengan nilai median

df_num.loc[df_num.oldpeak > 4 , "oldpeak"] = 0.8 #Mengganti nilai outliers dari variabel oldpeak dengan nilai median
df_num.loc[df_num.oldpeak < -2.4 , "oldpeak"] = 0.8 #Mengganti nilai outliers dari variabel oldpeak dengan nilai median
```

**One-hot encoder/membuat data dummy untuk variabel di dataset kategorik dan menggabungkannya dengan dataset numerik**

```python
cat_var = ['sex','cp','fbs','restecg','exng','slp','caa','thall'] #Membuat list variabel dari dataset numerik
df_cat = df_cat[cat_var] #Membaut dataset kategorik dengan nama variabel kategorik
df_cat = pd.get_dummies(df_cat,columns=cat_var) #Melakukan one-hot encoder
df = pd.concat([df_num,df['output'],df_cat],axis=1) #Menggabungkan dataet numerik dan kategorik
df #Menampilkan dataset
```

**Menentukan X sebagai variabel prediktor dan y sebagai variabel target**

```python
X = df.drop(['output'], axis = 1) #X sebagai variabel prediktor
y = df.output.values #y sebagai variabel target
```

**Membagi dataset menjadi data test dan train dengan rasio 2:8**

```python
from sklearn.model_selection import train_test_split #Mengimport split train test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) #Membagi dataset menjadi data train dan data test
```

**Standarisasi dataset**

```python
from sklearn.preprocessing import StandardScaler #Mengimport standarscaler
sc = StandardScaler() #standarscaler
X_train = sc.fit_transform(X_train) #Menstandarisasi dataset x_train
X_test = sc.transform(X_test) #Menstandarisasi dataset x_test
```

**Mengimport module dan menginstall keras tuner**

```python
import numpy as np #Mengimport numpy
import keras #Mengimpoort keras
import tensorflow as tf #Mengimport tenserflow
!pip install -q -U keras-tuner #Menginstall keras tuner

import keras_tuner as kt #Mengimport keras tuner
```

**Membagi data train dan data validasi**

```python
X_valid, X_train = X_train[:70], X_train[70:] #Membagi dataset x_train dan x_valid
y_valid, y_train = y_train[:70], y_train[70:] #Membagi datase y_train dan y_valid
```

**Membuat model klasifikasi dengan DNN**

```python
def build_model(hp): #Membuat fungsi model DNN
  model = keras.models.Sequential() #Membuat model DNN

  model.add(Dense(32, input_dim=30, activation="relu")) #layer pertama/input layer
  hp_layers = hp.Choice('layer', [3,4])  #jumlah layer
  hp_neurons = hp.Int('neuron', min_value = 10, max_value=100, step = 10)  #Jumlah neuron

  for layer in range(hp_layers): #for loop
    model.add(keras.layers.Dense(hp_neurons, activation = 'relu'))  #jumlah hidden layers

  model.add(keras.layers.Dense(2, activation = 'softmax'))  #layer terakhir/output layer
  optimizer = tf.keras.optimizers.Adam() #memasukan optimizer
  model.compile(optimizer='adam',loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy']) #meng-compile model DNN
    
  return model #Mengembalikan model DNN
```

**Hyperparameter tuning untuk model klasifikasi DNN**

```python
from keras.layers import Dense #Mengimport dense
tuner = kt.Hyperband(build_model, objective = 'val_accuracy',max_epochs = 10, directory = 'my_dir', project_name = 'my_tuner')

#early stopping
stop_early = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss',patience = 5)
#proses pencarian hyperparameter terbaik
tuner.search(X_train, y_train, epochs = 100, validation_data = [X_valid, y_valid], callbacks = [stop_early])
```

**Menampilkan jumlah layers dan neurons terbaik**

```python
best_hps=tuner.get_best_hyperparameters()[0] #Mengambil hyperparameter terbaik

print('best number of layers:',best_hps.get('layer')) #Menampilkan jumlah layers terbaik
print('best number of neurons:',best_hps.get('neuron')) #Menampilkan jumlah neuron terbaik
```

**Menjalankan model dengan parameter hasil dari hyperparameter tuning**

```python
model = tuner.hypermodel.build(best_hps) #Menjalankan model kalsifiksi DNN
```

**Fitting model**

```python
model.fit(X_train, y_train, epochs = 200, batch_size = 100,validation_split= 0.2) #Fitting model DNN
```

**Melakukan prediksi klasifikasi DNN**

```python
ypred=model.predict(X_test) #Melakukan prediksi 
ypred #Menampilkan hasil prediksi
```

**Mengembalikan indeks elemen maks dari suatu kolom/baris tertentu**

```python
y_pred=np.array([np.argmax(i) for i in ypred]) #Mengembalikan indeks elemen maks
```

**Evaluasi model**

```python
from sklearn.metrics import classification_report, accuracy_score #Mengimport klasifikasi report dan skor akurasi
print('Results for Binary Model') #Menampilkan tulisan hasil
print(accuracy_score(y_test, y_pred)) #Menampilkan skor akurasi
print(classification_report(y_test, y_pred)) #Menampilkan klasifikasi report
```
